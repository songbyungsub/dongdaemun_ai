```python
# 전처리 1번 : 레이블링(숫자 인코딩) train과 test30을 동시에 한다.
from sklearn.preprocessing import LabelEncoder
# 레이블 인코더 생성
cjk_encoder = LabelEncoder()
# 'Sex' 열에 레이블 인코딩 적용
X['Sex'] = cjk_encoder.fit_transform(X['Sex'])
test30['Sex'] = cjk_encoder.transform(test30['Sex'])
display(X.head())
display(test30.head())
```

```python
# 전처리 2번 : 스케일링(표준화) train과 test30을 동시에 한다.
from sklearn.preprocessing import StandardScaler
# 데이터 표준화 (Standardization)
scaler = StandardScaler()
X_train = scaler.fit_transform(X)
test30 = scaler.transform(test30)
display(X_train[0:3])
display(test30[0:3])
```

```python
print(type(X_train))
print(type(Y))
```

```python
# X_train, Y 학습
# 학습용 데이터와 테스트용 데이터로 분리
X_train, X_test, y_train, y_test = train_test_split(X_train, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
# 로지스틱 회귀 모델 초기화 및 학습
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
# 예측 및 평가
# 예측 수행
y_pred = model.predict(X_test)

# 모델 평가
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"모델 정확도: {accuracy}")
print("분류 보고서:\n", report)
```
