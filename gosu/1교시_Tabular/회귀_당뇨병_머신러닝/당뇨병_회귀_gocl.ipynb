```python
# Feature 셋과 Label 셋 분리
X = train.drop(columns=['target'])
Y = train['target']
display(X.head())
display(Y.head())
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_target, test_target = train_test_split(X, Y, test_size=0.3) #train과 test를 7:3의 비율로 나누었습니다. df를 그대로 적용
print(len(X), len(train_data), len(test_data)) #실제로 잘 나누어졌는지 확인

print("train ratio : {:.2f}".format(len(train_data)/len(X)))
print("test ratio : {:.2f}".format(len(test_data)/len(X)))
```

```python
from sklearn.linear_model import LinearRegression
multi_regressor = LinearRegression() #객체 생성
multi_regressor.fit(train_data, train_target) #학습
```

```python
#다중선형회귀 모델을 학습
```

```python
from sklearn.linear_model import LinearRegression
multi_regressor = LinearRegression() #객체 생성
multi_regressor.fit(train_data, train_target) #학습
```

```python
print(multi_regressor.intercept_)  #절편 (intercept): 모든 피쳐가 0인 경우 label 값
print(multi_regressor.coef_)       #회귀 계수 (coefficient): label에 영향을 미치는 값..
```

```python
multi_train_pred = multi_regressor.predict(train_data)
multi_test_pred = multi_regressor.predict(test_data)
```

```python
from sklearn.metrics import mean_squared_error

multi_train_mse = mean_squared_error(multi_train_pred, train_target)
multi_test_mse = mean_squared_error(multi_test_pred, test_target)
print(f"Multi Regression Train MSE is {multi_train_mse:.4f}")
print(f"Multi Regression Test MSE is {multi_test_mse:.4f}")
```

```python
# 이 값은 테스트 데이터에 대한 모델의 성능을 나타내는 지표
```

```python
from math import sqrt
print(f"Multi Regression Train RMSE is {sqrt(multi_train_mse):.4f}")
print(f"Multi Regression Test RMSE is {sqrt(multi_test_mse):.4f}")
#
```
